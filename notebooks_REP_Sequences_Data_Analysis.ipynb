```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
filepath = 'General_One_Line_All_Genomes_1.xlsx'
df_gen = pd.read_excel(filepath)
df_gen.head()
```

```python
mpl.style.use('ggplot')
```

```python
count, bin_edges = np.histogram(df_gen['General_Mean_Length'])
print(count) # frequency count
print(bin_edges) # bin ranges
```

```python
df_gen['General_Mean_Length'].plot(kind='hist', figsize=(8, 5))

# add a title to the histogram
plt.title('Mean Length of REP-like Sequence Found in Bacterial Genomes')
# add y-label
plt.ylabel('Count')
# add x-label
plt.xlabel('Base Length')
plt.grid(False)

plt.show()
```

```python
df_gen['General_Mean_Distance'].plot(kind='hist', figsize=(8, 5))

# add a title to the histogram
plt.title('Mean Distance between REP-like Sequences Found in Bacterial Genomes')
# add y-label
plt.ylabel('Count')
# add x-label
plt.xlabel('Base Length (Distance)')
plt.grid(False)

plt.show()
```

```python
df_gen['Average_Hits_Per_10kb'] = (df_gen['General_Hits'] / df_gen['Genome_Length']) * 10000
df_gen['Average_Gene_Hits_Per_10kb'] = (df_gen['Gene_Hits'] / df_gen['Genome_Length']) * 10000
df_gen['Average_Non_Gene_Hits_Per_10kb'] = (df_gen['Non_Gene_Hits'] / df_gen['Genome_Length']) * 10000
```

```python
df_gen['Average_Hits_Per_10kb'].plot(kind='hist', figsize=(8, 5), bins=np.arange(0,22,2), color = "blue")

plt.xticks(np.arange(0,20,2))
# add a title to the histogram
plt.title('Average REP-like Sequences Per 10kb')
# add y-label
plt.ylabel('Count')
# add x-label
plt.xlabel('Average Number of REP-like Sequences')
plt.grid(False)

plt.show()
```

```python
df_gen['Average_Gene_Hits_Per_10kb'].plot(kind='hist', figsize=(8, 5), bins=np.arange(0,10,1), color = "blue")

plt.xticks(np.arange(0,10,1))
# add a title to the histogram
plt.title('Average REP-like Sequences Within Genes Per 10kb')
# add y-label
plt.ylabel('Count')
# add x-label
plt.xlabel('Average Number of REP-like Sequences')
plt.grid(False)

plt.show()
```

```python
df_gen['Average_Non_Gene_Hits_Per_10kb'].plot(kind='hist', figsize=(8, 5), bins=np.arange(0,24,2), color = "blue")

plt.xticks(np.arange(0,24,2))
# add a title to the histogram
plt.title('Average REP-like Sequences in Non-Gene Regions Per 10kb')
# add y-label
plt.ylabel('Count')
# add x-label
plt.xlabel('Average Number of REP-like Sequences')
plt.grid(False)

plt.show()
```
